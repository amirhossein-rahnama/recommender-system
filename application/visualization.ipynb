```python

response_mapping = {
    'Sehr gut': 5,
    'Gut': 4,
    'Mäßig': 3,
    'Weniger': 2,
    'Weniger gut': 1,
    
}


questions = [
    'Wie neugierig sind Sie darauf, neue Dinge auszuprobieren und neue Ideen zu erkunden?',
    'Wie sehr schätzen und genießen Sie verschiedene Formen der Kunst, wie z. B. Musik, Malerei, Literatur...?',
    'Wenn Sie mit Veränderungen oder unbekannten Situationen konfrontiert werden, wie gut sind Sie darin, sich anzupassen und neue Erfahrungen zu akzeptieren?'
]


for question in questions:
    df1_all[question + '_numeric'] = df1_all[question].map(response_mapping)
    print(f"Unmapped or missing data in '{question}':", df1_all[df1_all[question + '_numeric'].isna()][question].unique())


numeric_questions = [q + '_numeric' for q in questions]
df1_all['mean_openness'] = df1_all[numeric_questions].mean(axis=1, skipna=True)


df1_all['status'] = df1_all['mean_openness'].apply(lambda x: 'open' if x > 3 else 'not open')


df1_all[['mean_openness', 'status']].head()
```
